```
import sklearn
from sklearn import datasets

# melakukan load iris datasets
iris = datasets.load_iris()
```

```
# membagi atribut dan target, dan di assign ke dalam variabel x, y
x = iris.data
y = iris.target
```

```
from sklearn import tree

# membuat model dengan decision tree classifier
clf = tree.DecisionTreeClassifier()
```

```
from sklearn.model_selection import cross_val_score

# mengevalluasi performa model dengan cross_val_score
# pada "cross_val_score()" memiliki 4 parameter, yaitu
# - parameter pertama = model Machine Learning-nya
# - parameter kedua   = atribut/field dari dataset
# - parameter ketiga  = label dari dataset
# - parameter keempat = jumlah fold yang akan dipakai pada cross validation
scores = cross_val_score(clf, x, y, cv = 5)
scores
```
