```
import sklearn
from sklearn import datasets
```

```
# melakukan load dataset dalam iris
iris = datasets.load_iris()
```

```
# memisahkan atribut dan label pada iris dataset
x = iris.data
y = iris.target
```

```
from sklearn.model_selection import train_test_split

# membagi dataset menjadi training dan testing
# pada "train_test_split()" memiliki 4 parameter, yaitu
# - parameter pertama = atribut dari train set
# - parameter kedua   = atribut dari test set
# - parameter ketiga  = persentase target dari train set
# - parameter keempat = target dari test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```
# menghitung panjang/jumlah data pada masing-masing direktori split
print(len(x))
print(len(x_test))
print(len(x_train))
print(len(y_test))
print(len(y_train))
```
